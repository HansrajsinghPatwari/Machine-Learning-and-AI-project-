```python
num_cols = df.select_dtypes(include=["float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns
```

```python
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

```

```python
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

```

```python
df = df.drop_duplicates()
```

```python
df['Age'] = df['Age'].astype("int64")
df['Gender'] = df['Gender'].astype("int64")
df['Family_History'] = df['Family_History'].astype("int64")
df['Diabetes'] = df['Diabetes'].astype("int64")
df['Alcohol_Consumption'] = df['Alcohol_Consumption'].astype("int64")
df['Physical_Activity'] = df['Physical_Activity'].astype("int64")
df['Stress_Level'] = df['Stress_Level'].astype("int64")
df['Heart_Attack_Risk'] = df['Heart_Attack_Risk'].astype("int64")
df['Smoking'] = df['Smoking'].astype("int64")
df['Hypertension'] = df['Hypertension'].astype("int64")

```

```python
X = df.drop('Heart_Attack_Risk', axis=1)
y = df['Heart_Attack_Risk']
```

```python
from imblearn.over_sampling import SMOTE

# create SMOTE object
smote = SMOTE(random_state=42)

# balance the data
X_new, y_new = smote.fit_resample(X, y)
```

```python
y_new.value_counts()
```

```python
from sklearn.model_selection import train_test_split

# split the data in train and test
X_train, X_test, y_train, y_test = train_test_split(
    X_new, y_new, train_size=0.8, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegressionCV


model = LogisticRegressionCV(max_iter=4000)

    # train the model
model.fit(X_train, y_train)


y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
